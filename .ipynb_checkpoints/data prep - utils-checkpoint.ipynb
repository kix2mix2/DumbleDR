```python
import os
import json
from PIL import Image
```

```python
## Create settings file
path = '/Users/morarica/Developer/DumbleDR/public/data/'

settings =[]
for folder in os.listdir(path):
    if folder.startswith('.'):
        continue
    print(folder)
    
    data_dict ={}
    data_dict["name"] = folder
    data_dict["description"] = ''
    data_dict["paths"] = []
    for file in os.listdir(path + folder +'/human_sample/'):
        if file.startswith('.'):
            continue
        #print(file)
        data_dict["paths"].append(file)
        
    
    
    
    settings.append(data_dict)

string = json.dumps(settings)  
settings       
        
```

```python
## Sample datasets
import pandas as pd
from sklearn.model_selection import train_test_split

path = '/Users/morarica/Developer/DumbleDR/public/data/'
sample = 300

for folder in os.listdir(path):
    if folder.startswith('.'):
        continue
    print(folder)
    
    
#     os.makedirs(path + folder +'/human_sample_mini/')
    idx = None
    print('==============')
    for file in os.listdir(path + folder +'/human_sample/'):
        if file.startswith('.'):
            continue
#         print(file)

        filename = path + folder +'/human_sample/' +file
        
        df = pd.read_csv(filename)
        print(file)
        
        n = df.shape[0]
        if n>sample and idx is None:
            print(df.shape)
            print('<===in===>')
            _, df1 = train_test_split(df, test_size=sample, stratify=df['labels'])
            idx = df1.index
            files = df1.filename
            
        df = df.loc[df['filename'].isin(files)]
        print(df.shape)
        df.to_csv(path + folder +'/human_sample_mini/' + file, index=False)
        print('------')
        
        
```
